```python
def binarize(value):
    if 18 <= value <= 25:
        return '18-25'
    elif 26 <= value <= 35:
        return '26-35'
    elif 36 <= value <= 49:
        return '36-49'
    elif 50 <= value <= 69:
        return '50-69'
    else:
        return '70-90'
df['age'] = df['age'].apply(binarize)
```

```python
def fraud_with_categories(column,width):
    non_fraud = df[df['is_fraud'] == 0][column].value_counts()
    fraud = df[df['is_fraud'] == 1][column].value_counts()

    plt.figure(figsize=(width,5))
    sns.barplot(non_fraud, color='blue')
    sns.barplot(fraud, color='red')
    sns.lineplot(non_fraud, label='Non-Fraud',marker='o' ,color='blue')
    sns.lineplot(fraud, label='Fraud',marker='o' ,color='red')
    plt.xticks(rotation=90)
    plt.show()
```

```python
fraud_with_categories('age',10)
```

```python
def categorize_time(time):
    hour = time.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Day'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply the function to the DataFrame
df['time_category'] = df['trans_date_trans_time'].dt.time.apply(categorize_time)
```

```python
df['trans_date'] = df['trans_date_trans_time'].dt.date
df['age_num'] = df['trans_date'].apply(lambda x: x.year) - df['dob'].dt.year
fraud_df = df[df['is_fraud'] == 1]

fig = px.scatter(
    fraud_df, x='trans_date', y='age_num',
    title='Age of Fraud Victims by Transaction Date',
    labels={'trans_date': 'Transaction Date', 'age_num': 'Age'},
    color='age_num', color_continuous_scale='Viridis'
)

fig.update_layout(
    title_font_size=20,
    xaxis_title='Transaction Date',
    yaxis_title='Age',
    xaxis_tickformat='%Y-%m-%d'  # Format date on x-axis
)
# Displaying the plot
fig.show()
```

```python
fraud_by_day_name = (pd.to_datetime(fraud_df['trans_date']).dt.day_name().value_counts().sort_values(ascending=False).reset_index().rename(columns={'trans_date': 'day' , 'count': 'fraud_count'}))
```

```python
fig = px.bar(fraud_by_day_name, x='day' , y='fraud_count',hover_data=['day' , 'fraud_count'], color='day', 
       labels={'fraud_count': 'Number of frauds', 'day': 'Week day'}, title='Number of Frauds by Day', height=600, width=600)

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.utils import resample
```

```python
# Separate majority and minority classes
df_majority = df[df['is_fraud'] == 0]
df_minority = df[df['is_fraud'] == 1]

# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # Sample with replacement
                                 n_samples=len(df_majority),    # Match number in majority class
                                 random_state=123) # Reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])

# Display new class counts
print(df_upsampled['is_fraud'].value_counts())
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x=MinMaxScaler().fit_transform(x)
# x=StandardScaler().fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf=RandomForestClassifier(max_depth=1200,n_estimators=120,random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
rf=LogisticRegression()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.metrics import accuracy_score
rf=ExtraTreesClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_pred,y_test)
```
